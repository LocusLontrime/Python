```python
# Locus Lontrime 36 366 98 989

# Импортируем необходимые библиотеки
import os
import matplotlib
import matplotlib.pyplot

# Получаем имя файла .csv
filename = os.path.join('C:\Games', 'data.csv')
filename
```

```python
# Импорт панд и загрузка данных из файла
import pandas as pd
data = pd.read_csv(filename)
data.head()
```

```python
# Данные об общей прибыли по всем товарам по каждому месяцу
total_profit = data['total_profit'].tolist()
total_profit
```

```python
# Данные об общих продажах всех товарам ( в ед.) по каждому месяцу
total_units = data['total_units'].tolist()
total_units
```

```python
# Далее получаем данные о продажах каждого товара в отдельности за каждый месяц
```

```python
facecream = data['facecream'].tolist()
facecream
```

```python
facewash = data['facewash'].tolist()
facewash
```

```python
toothpaste = data['toothpaste'].tolist()
toothpaste
```

```python
bathingsoap = data['bathingsoap'].tolist()
bathingsoap
```

```python
shampoo = data['shampoo'].tolist()
shampoo
```

```python
moisturizer = data['moisturizer'].tolist()
moisturizer
```

```python
# Удобным образом градуируем шкалу графика
Y_axis = [100000 * i for i in range(1, 6)]
Y_axis
```

```python
# Получаем лист значений по оси X (месяца)
months = data['month_number'].tolist()
months
```

```python
# Строим график прибылей по всем товарам за каждый месяц, он выбивается из общем картины по абсолютным значениям, мы покажем его отдельно

matplotlib.pyplot.plot(months, total_profit, label='profit')
matplotlib.pyplot.title('Profit')
matplotlib.pyplot.xlabel('Months')
matplotlib.pyplot.ylabel('Monthly profit')
matplotlib.pyplot.xticks(months)
matplotlib.pyplot.yticks(Y_axis)
matplotlib.pyplot.legend(loc='upper left')                         
matplotlib.pyplot.show()
```

```python
# Строим график продаж всех товаров за каждый месяц (в ед.), т.к. он также немного выбивается из общем картины по абсолютным значениям, мы покажем его отдельно

# Снова удобным образом градуируем шкалу графика
Y_axis = [10000 * i for i in range(1, 6)]
Y_axis

matplotlib.pyplot.plot(months, total_units, label='overall sales')
matplotlib.pyplot.title('Overall sales')
matplotlib.pyplot.xlabel('Months')
matplotlib.pyplot.ylabel('Monthly overall sales')
matplotlib.pyplot.xticks(months)
matplotlib.pyplot.yticks(Y_axis)
matplotlib.pyplot.legend(loc='upper left')
matplotlib.pyplot.show()
```

```python
# Прочитайте все данные о продажах продукта и покажите их с помощью многострочного график:
# данные по количественным продажам каждого из товаров отлично укладываются на один плот

# Устанавливаем удобный размер
matplotlib.pyplot.rcParams['figure.figsize'] = [16, 11]

# Название графика
matplotlib.pyplot.title('Sellings')

# Данные каждого товара
matplotlib.pyplot.plot(months, facecream, label='facecream')
matplotlib.pyplot.plot(months, facewash, label ='facewash')
matplotlib.pyplot.plot(months, toothpaste, label = 'toothpaste')
matplotlib.pyplot.plot(months, bathingsoap, label = 'bathingsoap')
matplotlib.pyplot.plot(months, shampoo, label = 'shampoo')
matplotlib.pyplot.plot(months, moisturizer, label = 'moisturizer')

# Помещаем легенду в левый верхний угол, чтобы она не мешала и не залезала на график
matplotlib.pyplot.legend(loc='upper left')
```

```python
# Прочитайте данные о продажах мыла для купания за все месяцы и покажите их с помощью столбчатой диаграммы.

# Задаем удобный размер диаграммы
matplotlib.pyplot.rcParams['figure.figsize'] = [10, 4]

X_axis = [i for i in range(1, 12 + 1)]

matplotlib.pyplot.bar([i for i in months], bathingsoap, label='bathingsoap')
matplotlib.pyplot.xticks(X_axis)
matplotlib.pyplot.show()
```

```python
# Импортируем reduce
from functools import reduce

# Считаем годовые продажи по каждому из товаров
facecream_year = reduce(lambda x, y: x + y, facecream, 0)
print(facecream_year)

facewash_year = reduce(lambda x, y: x + y, facewash, 0)
print(facewash_year)

toothpaste_year = reduce(lambda x, y: x + y, toothpaste, 0)
print(toothpaste_year)

bathingsoap_year = reduce(lambda x, y: x + y, bathingsoap, 0)
print(bathingsoap_year)

shampoo_year = reduce(lambda x, y: x + y, shampoo, 0)
print(shampoo_year)

moisturizer_year = reduce(lambda x, y: x + y, moisturizer, 0)
print(moisturizer_year)
```

```python
# Так будет виднее
matplotlib.pyplot.rcParams['figure.figsize'] = [16, 10]

# Задаем данные и лейблы
labels = ['facecream_year', 'facewash_year', 'toothpaste_year', 'bathingsoap_year', 'shampoo_year', 'moisturizer_year']
values = [facecream_year, facewash_year, toothpaste_year, bathingsoap_year, shampoo_year, moisturizer_year]

# Строим диаграмму
fig1, ax1 = matplotlib.pyplot.subplots()
ax1.pie(values, labels=labels, autopct='%1.2f%%')
ax1.axis('equal')
ax1.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))  # двигаем легенду подальше
matplotlib.pyplot.show()
```
